# Compute the drag coefficient

<p>The first step in calculating the Ekman currents is calculating **wind stress** from the wind speed obtained above. </p>

For this we use the relation:

$\tau = \rho C_D U_{10}^2$

Where $\rho$ is the density of air (1.3 $g.m^{-3}$), $U_{10}$ is the wind speed at 10 m above sea level, and $C_D$ is the drag coefficient.

```python
from IPython.core.display import HTML
HTML('<iframe src=http://mitgcm.org/pipermail/mitgcm-devel/attachments/20070815/d794d27e/attachment.pdf width=500 height=400></iframe>')
```

**Here we will use the parametrization from Yelland and Taylor [1996] for our calculation.**

```python
CD = np.ma.masked_all(wind.shape, wind.dtype)
cond1 = (3 <= wind) & (wind <= 6)
cond2 = (6 <= wind) & (wind <= 26)
CD[cond1] = (0.29 + (3.1/wind[cond1]) + 7.7/wind[cond1]**2)/1000.
CD[cond2] = (0.60 + 0.070*wind[cond2])/1000.
```

Ok we've got warning here! This is mainly due to **NaN** values.

# Calculated drag coefficient

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=80,
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon, lat))
# Remove z dimension and average over time
Z = np.ma.mean(CD[:,0], axis=0)  
colors = m.contourf(X, Y, Z)
cb = m.colorbar(colors)
m.drawcoastlines(linewidth=0.5)
ax.set_title('Drag coefficient Yelland&Taylor')
```

## Output the wind stress map

We can now calculate the wind stress:

```python
rho = 1.3
tau = rho * CD * wind**2
```

And we output the stress map.

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=80,
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon, lat))
Z = np.ma.mean(tau[:,0], axis=0)  # remove z dimension and average over time
colors = m.contourf(X, Y, Z)
cb = m.colorbar(colors)
cb.set_label("N m-2")
m.drawcoastlines(linewidth=0.5)
ax.set_title('Wind stress map')
```

# Get the wind stress vector

Once we have wind stress, we need to decompose it into **zonal (u)** and **meridional (v)** components in order to determine the Ekman transport. For this, we need the wind direction.

```python
direction = dataset.g_wind_from_direction.array[:]
direction = np.ma.masked_where(np.isnan(direction), direction)
print np.min(direction), np.max(direction)

# Direction is where wind comes from, change to direction it goes to
direction = 90 - direction

# convert to radians
direction *= np.pi/180.

tau_x = tau * np.ma.cos(direction)
tau_y = tau * np.ma.sin(direction)
```

## Plot the wind stress vector

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=80,
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon, lat))
U = np.ma.mean(tau_x[:,0], axis=0)  # remove z dimension and average over time
V = np.ma.mean(tau_y[:,0], axis=0)  # remove z dimension and average over time
m.quiver(X[::10,::10], Y[::10,::10], U[::10,::10], V[::10,::10])
m.drawcoastlines(linewidth=0.5)
ax.set_title('Wind stress vectors')
```

# Calculate the Ekman transport

From the zonal and meridional components we can calculate the Ekman transport using the following relations:

$M_x = \frac{\tau_y}{f}$

$M_y = -\frac{\tau_x}{f}$

Where $M_x$ is the zonal Ekman transport, $M_y$ is the meridional Ekman transport, and $f$ the Coriolis parameter and depends on the latitude:

$f = 2\Omega\sin(\theta)$

Where $\Omega$ is the angular velocity of the Earth and $\theta$ is latitude.

```python
Omega = 2*np.pi/(24*60*60)
f = 2*Omega*np.sin(lat*np.pi/180.)

Mx = tau_y/f[:,np.newaxis]
My = -tau_x/f[:,np.newaxis]
```

*Note:*

*Contrarily to common logic, North-South winds cause mass transport in the East-West direction. Indeed, the zonal component of the Ekman transport is related to the meridional component of the wind stress, and vice-versa.*

Here's how the zonal component looks like:

## Ekman transport plot

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=80,
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon, lat))
Z = np.ma.mean(Mx[:,0], axis=0)  # remove z dimension and average over time
colors = m.contourf(X, Y, Z)
cb = m.colorbar(colors)
cb.set_label("kg s-1 m-1")
m.drawcoastlines(linewidth=0.5)
ax.set_title('Zonal Ekman transport')
```

### Zonal component

Plot the zonal component of the Ekman transport:

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=80,
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon, lat))
Z = np.ma.mean(Mx[:,0], axis=0)  # remove z dimension and average over time
lon2, lat2 = np.meshgrid(lon, lat)
mask = abs(lat2) <= 10
Z = np.ma.masked_where(mask, Z)
colors = m.contourf(X, Y, Z, levels=np.linspace(-3000, 3000, 11), extend="both")
cb = m.colorbar(colors)
cb.set_label("kg s-1 m-1")
m.drawcoastlines(linewidth=0.5)
ax.set_title('Zonal Ekman transport')
```

## Meridional component

And plot the meridional component of the Ekman transport:

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=80,
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon, lat))
Z = np.ma.mean(My[:,0], axis=0)  # remove z dimension and average over time
lon2, lat2 = np.meshgrid(lon, lat)
mask = abs(lat2) <= 10
Z = np.ma.masked_where(mask, Z)
colors = m.contourf(X, Y, Z, levels=np.linspace(-3000, 3000, 11), extend="both")
cb = m.colorbar(colors)
cb.set_label("kg s-1 m-1")
m.drawcoastlines(linewidth=0.5)
ax.set_title('Meridional Ekman transport')
```
