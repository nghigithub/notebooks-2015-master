+ stratigraphic correlation (Hettinger, 2001, Ryder, 2002), 
+ petroleum reservoir analyses (Nelson and Bird, 2005), 
+ aquifer studies (Balch, 1988), and 
+ synthetic seismic profiles (Kulander and Ryder, 2005). 

Commonly, well logs are easier to visualize, manipulate, and interpret when available in a digital format.

## Log ASCII Standard

The **LAS** format (Log ASCII Standard) was introduced by the Canadian Well Logging Society in 1989 to standardize the organization of digital log curve information for personal computer users. It did this very successfully and the standard became popular worldwide. Since then, this file-format has been routinely used in the Oil&Gas and water well industries to store well log information. 

A single **LAS** file can only contain data for one well, but it can contain any number datasets (called "*curves*") from that well. Common curves found in a **LAS** file may include:

+ natural gamma, 
+ travel time, or 
+ resistivity logs.

For an in depth view of the file format have a look at this online [description](https://esd.halliburton.com/support/LSM/GGT/ProMAXSuite/ProMAX/5000/5000_8/Help/promax/las_overview.pdf).

# LAS file visualisation

From the <code>keys()</code> function above we saw that the file contains 7 curves: DEPT, DNCP, DFCP, GCPS, DENF, DENN, GAMM.


```python
for i, curve in enumerate(LAS.curves[0:]):
    print i,' - Curve Name: ',curve.mnemonic
```

```python
# Figure size
fig = plt.figure(figsize=(12, 5))

# Specifies the geometry of the grid of each subplot
gs = gridspec.GridSpec(1, len(LAS.curves)-1, wspace=0.3)

# Loop from the data curve starting from the DNCP up to GAMM
for i, curve in enumerate(LAS.curves[1:]):
    
    # Subplot definition
    ax = fig.add_subplot(gs[i])
    
    # Plot each curves against well depth
    ax.plot(curve.data, LAS["DEPT"])
    
    # Set plot title
    ax.set_title(curve.mnemonic)
    
    # Write the y axis on the first plot
    if i > 0:
        ax.set_yticklabels([])
    
    # Define x axis labels and limit
    ax.set_xticklabels(["%.0f" % xi if xi in ax.get_xlim() else "" for xi in ax.get_xticks()], rotation=30)
```

# Multi-scale log attributes (just for your information...)

Using IPython and the LAS_reader package it is possible to **explore stratigraphic stacking patterns in scale space**. It's a little like spectral decomposition for 1D data. 
<br/>
The idea is to average a log at a range of window sizes, increasing exponentially.  This could for example being use to help **train a facies classifier**.
<br/>
## First moment integration

```python
# Loading the Panuke well
well = las_reader.read('LASdata/Panuke_B-90.las')

# Setting up variables
DT = well['DT']
RHOB = well['RHOB']
GR = well['GR']
z = well['DEPTH']
dz = well.well['STEP'].value

# Defining decomposition
start = 1000
stop = 25000
n = (stop-start)/1000
GR = GR[start:stop]
z = z[start:stop]
ncols = 50

# Create first moment function
def moment1(GR, dz, moment=1, max_window= 820, ncols=50):
    """
    GR is log
    dz is sample rate (m)
    max_window = 1640 (m)
    moment: is the nth order moment to calculate
    """
    w = max_window / dz  # integer with log base 2 max num windows
    m = np.ones( (len(GR), ncols))
    num = int(np.log2(w))
    
    # first moment 
    for j, window in enumerate(np.logspace(0, num, ncols,
                                endpoint=True, base=2)):
        print j, window
        for i in np.arange(int(window),len(GR)-int(window)):
            m[i,j] = np.mean(GR[ i-window: i+window+1 ] )
        m[:,j] /= np.amax(m[:,j])         
    
    return m

GRimg = np.repeat(np.expand_dims(GR,1),ncols,1)

# Perform first moment calculation over decomposition range
m1 = moment1(GR,dz, ncols)
print "finished calculating first Moment"
```

## Plot calculated attribute

```python
fig  = plt.figure(figsize = [4,12], facecolor='white')
ax1 = fig.add_axes([0.05, 0.1, 0.4, 0.85])
ax2 = fig.add_axes([0.7, 0.1, 0.4, 0.85])

# plot log
ax1.plot(GR[::n],z[::-n],'k' )
#ax1.axis('off')
ax1.axis('tight')

ax2.imshow(GRimg, cmap='gist_earth_r')

scGR = ncols*GR/np.amax(GR)

maxpt = np.argmax(np.transpose(m1),axis=0)
maxpt = ncols*maxpt/(2**13)

# with log on top
ax2.plot(scGR,np.arange(len(scGR)),'k')

# make the color fill clip
ax2.fill_betweenx(np.arange(len(scGR)), scGR, x2=100, color='white')
#ax2.axis('off')
ax2.axis('tight')
```
