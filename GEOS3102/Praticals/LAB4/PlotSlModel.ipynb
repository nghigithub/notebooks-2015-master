```python
import h5py
import numpy as np
import matplotlib.mlab as ml
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ETO
import scipy.interpolate as scint

%matplotlib inline
```

```python
t_nosl = np.array([0.,500000])
nosl = np.array([0.,0.])

t_rise = np.array([0.,180000.,200000,280000,300000,350000,370000,400000,420000,500000])
sl_rise = np.array([-150.,-150.,-100.,-100.,-50.,-50.,0.,0.,50,50])

t_fall = np.array([0.,180000.,200000,280000,300000,320000,380000,400000,420000,500000])
sl_fall = np.array([0.,0.,-75.,-75.,-100.,-100.,-150.,-150.,-200.,-200.])
```

```python
pty = 12500.
fig = plt.figure(figsize = (5,10))
plt.rc("font", size=12)
#
ax1 = fig.add_subplot(1, 1, 1) 
minX = -200.
maxX = 50.
minZ = t_rise.min()
maxZ = t_rise.max()

plt.plot(nosl,t_nosl,'-',color='#C0C0C0',linewidth=3,label='model 1 no fluctuations')

plt.plot(sl_fall,t_fall,'o-',color='#6666FF',linewidth=2,label='model 2 sea level fall')

plt.plot(sl_rise,t_rise,'o-',color='#FF3399',linewidth=2,label='model 3 sea level rise')


axes = plt.gca()
plt.xlim( minX-10, maxX+10 )
plt.ylim( minZ, maxZ )
plt.grid(True)
#plt.title('Sea level fluctuations for each model',fontsize=16)
plt.xlabel('Sea level (m)',fontsize=14)
plt.ylabel('Time (years)',fontsize=14)
plt.legend(loc=4, fontsize=10)
```
